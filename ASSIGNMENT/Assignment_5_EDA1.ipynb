```python
# Outlier Detection: User defined Function to calculate Upper Extreme and Lower Extreme value
def outlier_detection(data,colname):
  q1 = data[colname].quantile(0.25)
  q3 = data[colname].quantile(0.75)
  iqr = q3 - q1

  upper_extreme = q3 + (1.5 * iqr)
  lower_extreme = q1 - (1.5 * iqr)

  return lower_extreme, upper_extreme,q1,q3
```

```python
# Histogram of numerical variables
df.hist(figsize=(15, 10), bins=20)
plt.suptitle('Histograms of Numerical Variables')
plt.show()
```

```python
# Boxplot of numerical variables
plt.figure(figsize=(8,8))
sns.boxplot(data=df)
plt.xticks(rotation=50)
plt.title('Boxplot of Numerical Variables')
plt.show()
```

```python
# Let's create categories: Low, Normal, High
bins = [0, 110, 160, np.inf]  # Example thresholds
labels = ['Low', 'Normal', 'High']
df['LB_Category'] = pd.cut(df['LB'], bins=bins, labels=labels)

# Display frequency counts for the new categorical variable
print("Frequency of LB Categories:\n", df['LB_Category'].value_counts())
```

```python
# Pie chart for 'LB_Category'
plt.figure(figsize=(5, 8))
df['LB_Category'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis'), startangle=90)
plt.title('Distribution of Baseline Fetal Heart Rate (LB) Categories')
plt.ylabel('')  # Hide y-label for pie chart
plt.show()
```

```python
# Scatter plot correlation heatmap
# Scatter plot between 'LB' and 'AC'
plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='LB', y='AC')
plt.title('Scatter plot between LB and AC')
```

```python
#  correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt='.2f') # Select only numerical columns for correlation calculation
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pair plot
sns.pairplot(df)
plt.title('Pair Plot of Variables')
plt.show()
```

```python
# Violin plot for a specific variable
plt.figure(figsize=(8, 6))
sns.violinplot(x='UC', y='LB', data=df)
plt.title('Violin Plot of UC vs LB')
plt.show()
```

# 4**.	Pattern Recognition and Insights:**

笳十tIdentify any correlations between variables and discuss their potential implications.

笳十tLook for trends or patterns over time if temporal data is available.


```python
# Calculate correlations for numerical columns only
correlations = df.select_dtypes(include=np.number).corr()
correlations
```

```python
# Select the columns LB and AC
selected_columns = df[['LB', 'AC']]

# Calculate the correlation matrix for LB and AC
correlation_matrix = selected_columns.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix for LB and AC')
plt.show()
```

The correlation matrix will provide a single value representing the correlation between LB and AC. This value can range from -1 to 1:

A value close to 1 indicates a strong positive correlation, meaning that as LB increases, AC also tends to increase. A value close to -1 indicates a strong negative correlation, meaning that as LB increases, AC tends to decrease. A value close to 0 indicates no significant linear relationship between LB and AC.

The correlation matrix of LB and AC is Positive Correlation (0.26), The correlation value of 0.26 indicates a moderate to strong positive relationship between baseline FHR and accelerations.
