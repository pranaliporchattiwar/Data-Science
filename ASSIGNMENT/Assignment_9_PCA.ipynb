# **2: Dimensionality Reduction with PCA**

1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.
2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
3.	Transform the original dataset into the principal components.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Standardize the features
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df
```

```python
# PCA
pca = PCA()
pca.fit(scaled_df)
```

```python
# Scree plot to determine the number of principal components
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.show()
```

```python
# Transform the dataset using PCA
optimal_components = 2  # Based on the scree plot or explained variance
pca = PCA(n_components=optimal_components)
pca_df = pca.fit_transform(scaled_df)
```

```python
# Check the transformed dataset
print(pd.DataFrame(pca_df).head())
```

# **3: Clustering with Original Data**

1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.
2.	Visualize the clustering results using appropriate plots.
3.	Evaluate the clustering performance using metrics such as silhouette score or Davies–Bouldin index



```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Apply K-means to the original dataset
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust n_clusters based on your dataset
kmeans_labels = kmeans.fit_predict(df)
```

```python
# Visualize the clustering results
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df.iloc[:, 0], y=df.iloc[:, 1], hue=kmeans_labels, palette='viridis')
plt.title('Clustering on Original Data')
plt.show()
```

```python
# Evaluate clustering performance
silhouette = silhouette_score(df, kmeans_labels)
db_score = davies_bouldin_score(df, kmeans_labels)
print(f"Silhouette Score (Original Data): {silhouette}")
print(f"Davies-Bouldin Index (Original Data): {db_score}")
```

# **4: Clustering with PCA Data**

1.	Apply the same clustering algorithm to the PCA-transformed dataset.
2.	Visualize the clustering results obtained from PCA-transformed data.
3.	Compare the clustering results from PCA-transformed data with those from the original dataset.


```python
# Apply K-means to the PCA-transformed dataset
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca_labels = kmeans_pca.fit_predict(pca_df)
```

```python
# Visualize the clustering results from PCA-transformed data
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_df[:, 0], y=pca_df[:, 1], hue=kmeans_pca_labels, palette='viridis')
plt.title('Clustering on PCA-transformed Data')
plt.show()
```

```python
# Evaluate clustering performance
silhouette_pca = silhouette_score(pca_df, kmeans_pca_labels)
db_score_pca = davies_bouldin_score(pca_df, kmeans_pca_labels)
print(f"Silhouette Score (PCA Data): {silhouette_pca}")
print(f"Davies-Bouldin Index (PCA Data): {db_score_pca}")
```
